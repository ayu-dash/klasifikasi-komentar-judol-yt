```python
from googleapiclient.discovery import build
import time
import pandas as pd
```

```python
api_key = ""
video_id = "YZ4N8jH5R_M" 
video_ids  = ["YZ4N8jH5R_M", "s9OU_mLo-KU", "1msXOdJcG9s", "rkoymgMW-8M", "yY76VsIplzo", "JpaK8OhL4FI", "UnVihN2_M2U", "GHbSjBdMB8E", "4k6rzuj0bWI", "FpSJFqYaRb8", "dXtcUtRJO0g"]
```

```python
youtube = build("youtube", "v3", developerKey=api_key)

# ======================================
# AMBIL KOMENTAR DARI SEMUA VIDEO
# ======================================
comments = []

for vid in video_ids:
    print(f"ğŸ” Mengambil komentar dari video: {vid}")
    next_page_token = None

    while True:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=vid,
            textFormat="plainText",
            maxResults=100,
            pageToken=next_page_token
        )
        response = request.execute()

        for item in response["items"]:
            snippet = item["snippet"]["topLevelComment"]["snippet"]
            comments.append({
                "comment_id": item["id"],
                "video_id": snippet["videoId"],
                "author": snippet["authorDisplayName"],
                "comment_text": snippet["textOriginal"],
                "published_at": snippet["publishedAt"],
                "like_count": snippet.get("likeCount", 0)
            })

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break
            
        time.sleep(1)
```

```python
df = pd.DataFrame(comments)
```

```python
df.to_csv("komentar.csv")
```
